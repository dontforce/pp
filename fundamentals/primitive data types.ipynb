```python
#### FUNDAMENTALS #####

# Primitive Data types

# string
x = 'lebron > jordan'

# common string methods

print( x.upper() )
print( x.lower() ) # upper/lower case

# splitting the word on a character
split = x.rsplit( ' > ', 2 ) # this is now a list of the split up string
print( "the split list: {}".format( split ) )
lebron = split[ 0 ]
jordan = split[ 1 ]
print( "{} > {}".format( lebron, jordan ) )

# 

```

```python
import sys

# The numbers: int, double, float, long, short. 

# int: no decimals. 4 bytes of storage. 1 byte = 8 bits, so an int is 32 bits. So the highest number you can store
# in an int is -2^31 < 0 > 2^31, -2,147,483,648 to 2,147,483,648
x = 31
y = 2

print( "x = {}, y = {}".format( x, y ) ) # best way to print when you need to pass in multiple variables

# operations: PEMDAS ( ** * / + - ), MOD: %

# +=, -=
x += y # operation + assignment
print( 'x += y (same as x = x + y): ', x )

x -= y
print( 'x -= y: ', x )

# * / ^
x = y * x # 62
print( 'y * x = ', x )
x = x / y # 31
print( 'x / y = ', x ) # dividing turns an int into a double

x = int( x ) # you can `cast` (transform one data type to another) the double back into an int
print( "after cast: ", x )

# or instead of casting, you can do integer division (double //) (dont need to cast)
x = ( y * x ) // y
print( "integer division: ", x )

# exponents
y = y ** x
print( "y ** x = ", y )

# could also do this
y = 2 # reset
y **= x
print( "This is also the largest int value: ", y )

# lets try to add one to the largest int value
y += 1
print( "Trying to overflow the integer... ", y )

# This works because in the background, python turns the number into a long (which is much bigger than an int)
# in other programming languages like java or C, it wouldn't let you do it and throw an Error


# FLOAT, DECIMAL: floating point number (decimals). 32 bits, 4 bytes.

x = 2
x = float(x)
print( "float: ", x )

x = 23243423.0
print( type(x) ) # all decimal numbers are float by default in python

x = 21.01234567890123456
print( type(x), x ) # float is not precise with the decimal places. only accurate up to 15 decimal places

x = 2147483648.0123456789012345
print( type(x), x )


# DOUBLE: same as float, but 64 bits (8 bytes). Much larger than both int and double. It's also more precise than float.

# However, double isn't a primitive data type for python, you can use the Decimal library to use them though

# in every other major language (like java), its one of the most used data types.



# LONG: A much larger int. 64 bits, 8 bytes. No decimals. Not a primitive data type for python. ( useful in java, not python)

# Short: A much smaller int. 16 bits, 2 bytes. No decimals. Not a primitive data type for python. ( never use em )
```
